```python
import pandas as pd
import numpy as np
import matplotlib as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("netflix_titles.csv")
df.head()
#df.columns
```

```python

df['listed_in'].unique
```

```python
list1=[]
for l in df['listed_in']:
    l=l.split(",")
    list1.append(l[0])
df['list1']=list1
df.columns
del list
```

```python
del df['list']
df.head()
```

```python
df=df.rename(columns={'list':'list1'},inplace=False)
df.head()
```

```python
features=['director','cast','rating','list1']
```

```python
for fe in features:
    df[fe]=df[fe].fillna(' ')
```

```python
def combine(row):
    return row['director']+" "+row['cast']+" "+row['rating']+" "+row["list1"]
df['combined_feat']=df.apply(combine,axis=1)
#print(df['combined_feat'])
```

```python
#creating a count matrix from this combined column
cv=CountVectorizer()
count_mat=cv.fit_transform(df['combined_feat'])
#print(count_mat)
```

```python
cosine_similarities=cosine_similarity(count_mat)
```

```python
movie_liked='Zozo'
```

```python
def get_title(index):
    return df[df.index==index]["title"].values[0]
def get_index(title):
    return df[df.title==title]["row_num"].values[0]
```

```python
movie_index=get_index(movie_liked)
print(movie_index)
print(cosine_similarities)
```

```python
similar_movies=list(enumerate(cosine_similarities[movie_index]))
sorted_similar_movies=sorted(similar_movies,key= lambda x:x[1],reverse=True)
```

```python
#data=df.iloc[0:-1]
#data.colump
i=0
for ele in sorted_similar_movies:
    print(get_title(ele[0]))
    i=i+1
    if i>50:
        break
```
